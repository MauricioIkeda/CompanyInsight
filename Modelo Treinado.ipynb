# Importando As Bibliotecas

```python
import spacy
```

# Carregando o Modelo 1.0

```python
nlp_ner = spacy.load("model-last") 
```

# Testando Modelo 1.0 Com Reclamações

```python
doc = nlp_ner('''o Celular da Positivo não está funcionando''')

spacy.displacy.render(doc, style="ent", jupyter=True)
```

```python
doc = nlp_ner('''Tenho duas lâmpadas inteligente Positivo, são excelentes mas recentemente uma delas começou a ficar offline, vi que é um problema recorrente da empresa, e gostaria de uma solução, já reiniciei, desliguei internet, procurei tutoriais, mas acredito que o problema esteja no produto, já que a outra está funcionando perfeitamente. Desde já agradeço!''')

spacy.displacy.render(doc, style="ent", jupyter=True)
```
